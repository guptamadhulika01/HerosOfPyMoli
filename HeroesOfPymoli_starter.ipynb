## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# create a new list with desired columns & display the head 
popularItems_df = pd.DataFrame(purchase_df[["Item ID","Item Name", "Price"]])
popularItems_df.set_index('Item ID', inplace=True)
popularItems_df.head()
 
```

```python
#groupby Item ID and Item Name, obtain Purchase Count; sort by descending order
groupedItems = pd.DataFrame(popularItems_df.groupby(['Item ID','Item Name']).Price.count())
groupedItems = groupedItems.rename(columns = {"Item ID":" Item ID", "Item Name":"Most Profitable Items", "Price": "Count Of Items "})
#descendedgroupedItems = pd.DataFrame(groupedItems.sort_values('Price',ascending = False))
groupedItems.head()
```

```python
#Merge DataFrame on outer Join to add columns
#>head() giving silly results as the Item ID is repeated several Item. 
df_outer = pd.merge(popularItems_df, groupedItems, on='Item ID', how='outer')


```

```python
#Total Revenue grouped by Item ID and Item Name
TotalRevenue= pd.DataFrame(popularItems_df.groupby(['Item ID','Item Name']).sum())
TotalRevenue = TotalRevenue.rename(columns = {"Item ID":" Item ID", "Item Name":"Most Profitable Items", "Price": "Total Revenue"})
TotalRevenue.head()
```

```python
#Merge DataFrame on outer Join to add columns
#>head() giving silly results as the Item ID is repeated several Item. 
#delete duplicate rows form the DataFrame
df_outer1 = pd.merge(df_outer, TotalRevenue, on='Item ID', how='outer')
DropDuplicates = pd.DataFrame(df_outer1.drop_duplicates())
DropDuplicates.head()
```

```python
#Sort Purchase value in descending order

sortedTotalPurchaseValue = pd.DataFrame(DropDuplicates.sort_values('Total Revenue', ascending=False))
sortedTotalPurchaseValue.head()
```
